```python
import rescomp
import numpy as np
np.random.seed(1)

import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
#Simulation Parameters:
dt = 0.1
T = 200 # 5000 # training time
time_steps = int(T/dt)

# KS Parameters:
L = 32*np.pi
Q = 256

x = L * np.transpose(np.conj(np.arange(1, Q + 1))) / Q
starting_point = np.cos(2 * np.pi * x / L) * (1 + np.sin(2 * np.pi * x / L))
# plt.plot(starting_point)

sim_data = rescomp.simulations._kuramoto_sivashinsky_Bhatt(dimensions = Q, system_size = L, dt = dt, time_steps = time_steps, starting_point = starting_point, alpha = 1, beta = 1)
to_plot = sim_data

# ax = plt.gca()
# ax.imshow(to_plot.T, aspect='auto')
```

```python
sim_data_FFT = rescomp.simulations._kuramoto_sivashinsky(dimensions = Q, system_size = L, dt = dt, time_steps = time_steps, starting_point = starting_point, eps = 0)
```

```python
fig, axs = plt.subplots(3, 1, sharex="all", figsize=(9, 6),
                            constrained_layout=True, dpi=300)

ax = axs[0]
im = ax.imshow(sim_data.T, aspect='auto')
ax.set_title("New")

ax = axs[1]
im = ax.imshow(sim_data_FFT.T, aspect='auto')
ax.set_title("FFT")

ax = axs[2]
im = ax.imshow((sim_data - sim_data_FFT[1:, :]).T, aspect='auto')
ax.set_title("Dif")

fig.colorbar(im, ax=axs)
```
